The Top 5 brand 

```python
autos['brand'].value_counts(normalize = True)
```

```python
dict_mean_price = {}
for brnd in autos['brand'].value_counts().index[:17]:
    mean = autos.loc[autos['brand'] == brnd, 'price'].mean()
    dict_mean_price[brnd] = mean
sorted_dict_mean = sorted(dict_mean_price.items())
```

```python
print(dict_mean_price)
print(sorted_dict_mean)
```

```python
type(sorted_dict_mean)
```

```python
top6_mean_price = {}
top6_mileage = {}
for brnd in sorted_dict_mean[:6]:
    brand = brnd [0]
    average_price = autos.loc[autos['brand'] == brand, 'price'].mean()
    average_mileage = autos.loc[autos['brand'] == brand, 'odometer_km'].mean()
    top6_mean_price[brand] = average_price
    top6_mileage[brand] = average_mileage
print(top6_mean_price)
print(top6_mileage)
```

```python
series_mean_price = pd.Series(top6_mean_price)
series_mileage = pd.Series(top6_mileage)
```

```python
print(series_mileage)
```

```python
df = pd.DataFrame(series_mean_price, columns = ['mean_price'])
```
